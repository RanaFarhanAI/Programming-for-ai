```python
capacity_a = 4
capacity_b = 3
target = 2

visited_states = set()

def dfs(a, b):
    
    if (a, b) in visited_states:
        return False
    
    visited_states.add((a, b))
    
    print(f"State Reached: ({a}, {b})")
    
    if a == target or b == target:
        print("\nGoal State Found!")
        return True
    
    
    moves = []

    moves.append((capacity_a, b, "Fill Jug A"))
    moves.append((a, capacity_b, "Fill Jug B"))
    moves.append((0, b, "Empty Jug A"))
    moves.append((a, 0, "Empty Jug B"))
    
    pour_to_b = min(a, capacity_b - b)
    moves.append((a - pour_to_b, b + pour_to_b, "Transfer A → B"))
    
    pour_to_a = min(b, capacity_a - a)
    moves.append((a + pour_to_a, b - pour_to_a, "Transfer B → A"))
    
    for new_a, new_b, step in moves:
        if dfs(new_a, new_b):
            print("Action Taken:", step)
            return True
    
    return False


print("----- Depth First Search Started -----\n")
dfs(0, 0)
```
